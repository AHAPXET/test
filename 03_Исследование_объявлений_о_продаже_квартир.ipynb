```python
#По порядку проверяю на наличие различных нестыковок или неявных дублей
print(df['total_images'].sort_values().unique())
print()
print(df['last_price'].sort_values().unique())
print()
print(df['total_area'].sort_values().unique())
print()
display(df.query('total_area >= 400').sort_values(by='total_area'))
print()
print(df['first_day_exposition'].sort_values().unique())
#Время в большинстве указывается без часов, при переводе в дату можно оставить только год, месяц, день
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format = '%Y-%m-%d')




```

```python
print('Количество комнат')
print(df['rooms'].sort_values().unique())
print()
print('Высота потолков')
print(df['ceiling_height'].sort_values().unique())
#есть странные значения высоты потолков после 10 метров кажется странные значения, нужно проверять графиками
print()
print('Максимальные этажи')
print(df['floors_total'].sort_values().unique())
#df['floors_total'] = df['floors_total'].astype('int') - вызывается ошибка нужно разобраться, скорее всего из-за нулевых значений
#странные значения по максимальным этажам, выше 33 этажа надо проверить, возможно выбросы

print()
print('Жилая площадь')
print(df['living_area'].sort_values().unique())
```

```python
print()
print('Этажи в объявлении')
print(df['floor'].sort_values().unique())
#вполне возможно, что 33 этаж на последнем, надо проверить
print()
print('Объявления выше 25 этажа')
#display(df.query('floor >= 25')) #посмотрим что там происходит (не работает локально)
display(df.query("floor.astype('int') >= 25")) #посмотрим что там происходит
print()
print('Этаж в объявления выше максимального')
display(df.query("floor.astype('int') > floors_total")) #проверим есть ли логические ошибки
print()
print('Максимальны этаж выше 40')
display(df.query('floors_total >= 40')) 
#два объекта в разных регионах, очень странно возможно ошибка. 60 - 6 этажей всего(что редко), 
# а 52 - 22 или 32 который чаще встречается
```

```python
print()
print('Варианты апартаментов')
print(df['is_apartment'].sort_values().unique())
print()
print('Смотрим дату начала заполнения апартов')
display(df[~df['is_apartment'].isna()].sort_values(by='first_day_exposition'))
print()
print('Смотрим даты незаполненых апартов')
display(df[df['is_apartment'].isna()].sort_values(by='first_day_exposition'))
#явной зависимости нет, можем поставить вместо пустых значений False т.к. 
#пользователь мог просто не выбрать Нет, когда публиковал объявление
df['is_apartment'] = df['is_apartment'].fillna(value=False)

```

```python
print(df['balcony'].sort_values().unique())
df['balcony']=df['balcony'].fillna(0).astype(int)
#поставим 0 если не заполнено значение наличия балкона и переведем в int

print()
print(df[df['days_exposition'].isna()])
#мы не знаем дату снятия с публицации, возможно объявление еще весит, поэтому предположим, 
#что там где ячейки пустые объявление не снято, а значит можно поставить 0
df['days_exposition'] = df['days_exposition'].fillna(0).astype(int)

print()
print(df['locality_name'].sort_values().unique())
print(df['locality_name'].count())
#присутствуют неявные повторы к примеру "городской посёлок Янино-1" и "городской поселок Янино-1"
#всего 23650
#df['locality_name'] = df['locality_name'].str.lower()
#привести к нижнему регистру, не самая лучшая идея т.к. есть города и они должны начинаться с Большой буквы
```

```python
# подставим в не заполненные парки 0 так как мользователь мог не выбрать 0, так как рядом их нет. 
df['parks_around3000'] = df['parks_around3000'].fillna(value=0).astype(int)
df['ponds_around3000'] = df['ponds_around3000'].fillna(value=0).astype(int)

display(df[df['parks_nearest'].isna() & (df['parks_around3000'] == 0)])
# 15620 строк
display(df[df['ponds_nearest'].isna() & (df['ponds_around3000'] == 0)])
# 14589 строк

display(df[~df['parks_nearest'].isna() & (df['parks_around3000'] == 0)])
#Есть парки которые дальше 3000 метров, заполнять пропуски в parks_nearest нулем (0), не совсем корректно, 
#заполним пропуски медианными значениями по группам

df['parks_nearest'] = df['parks_nearest'].fillna(df.groupby('parks_around3000')['parks_nearest'].transform("median"))
#df['ponds_nearest'] = df['ponds_nearest'].fillna(df.groupby('ponds_around3000')['ponds_nearest'].transform("median"))
#Через группировку не работает, просто подставим 0
df['ponds_nearest'] = df['ponds_nearest'].fillna(value=0)
```

```python
#Высота потолков заменяет на медианное значение
df['ceiling_height'].fillna(df['ceiling_height'].median(), inplace=True)

df.isna().mean()*100
```

```python
#т.к. пропуски в locality_name и floors_total составляют меннее 1% удалим незаполненные
df.dropna(subset=['locality_name'], inplace=True)
df.dropna(subset=['floors_total'], inplace=True)
#Переведем значение этажа в целое
df['floors_total']=df['floors_total'].astype(int)

df.isna().mean()*100
```

```python
#не заполненные строки с жилой площадью и площадью кухни заполним медианным значением разбитым по группам
df['living_area'] = df['living_area'].fillna(df.groupby('rooms')['living_area'].transform("median"))
df['kitchen_area'] = df['kitchen_area'].fillna(df.groupby('living_area')['kitchen_area'].transform("median"))
df.isna().mean()*100
```

```python
#не заполненные строки с площадью кухни (0.26%) заполним через разность общей и жилой
df['kitchen_area'] = df['kitchen_area'].fillna(df['total_area'] - df['living_area'])

#Проверим оставшиеся пропуски
df.isna().mean()*100
```

* этаж квартиры; варианты — первый, последний, другой;
* соотношение жилой и общей площади, а также отношение площади кухни к общей.

```python
# Создаю новый столбец для цены квадратного метра -  prices_sq_meter
df['prices_sq_meter'] = df['last_price'] / df['total_area']
df['prices_sq_meter'] = df['prices_sq_meter'].astype(int)

# Столбцы дня недели, месяца и года
df['weekday'] = df['first_day_exposition'].dt.weekday
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year


# Создаём функцию для заполнения нового столбца с типом этажа
def floor_group(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'

#категоризуем этажи с помощью функции  
df['floor_category'] = df.apply(floor_group, axis = 1)

#добавим столбец соотношения жилой площади к общей
df['useful_area_ratio'] = df['living_area'] / df['total_area']

#добавим столбец отношения площади кухни к общей
df['kitchen_area_ratio'] = df['kitchen_area'] / df['total_area']

df
```
